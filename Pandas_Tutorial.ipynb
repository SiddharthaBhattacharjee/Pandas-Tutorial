#### Iterating through each Row

```python
for index,row in pokemon.iterrows():
    print(index,"==>\n",row) # prints row as an object , i.e. index is integer, row is object
```

```python
for index,row in pokemon.iterrows():
    if row['Generation'] != 1:
        break; # only allow gen 1 pokemon
    print(index,"==>\n",row[['Name','Type 1','Type 2']]) #print only name and types
```

#### Condition based selection
Its some what similar to database queries

```python
pokemon.loc[pokemon['Type 1'] == "Fire"] # select all pokemon whose type 1 is fire
```

```python
pokemon.loc[(pokemon['Type 1'] == "Fire") | (pokemon['Type 2'] == "Fire")] #selects all fire type pokemon i.e. either type should be fire
```

```python
pokemon.loc[((pokemon['Type 1'] == "Fire") | (pokemon['Type 2'] == "Fire")) & (pokemon['Legendary'])] # Select all fire type legendary pokemon
```

```python
pokemon.loc[((pokemon['Type 1'] == "Psychic") | (pokemon['Type 2'] == "Psychic")) & (pokemon['Legendary'])]['Name']
#select only the names of all psychic type legendary pokemon
```

```python
pokemon.loc[pokemon['Name'] == "Mew"] # search data of a specific pokemon by name
```

```python
pokemon.loc[pokemon['Name'] == "Snorlax"]['HP'] # find a specific value for a specific pokemon
```

## Practice Question:Find number of pokemon with above average speed for each type
(Try it yourself before expanding the solution)

```python
# find the types of pokemon
pd.unique(pokemon['Type 1'])
```

```python
#Solution
import numpy as np
avg_speed = np.average(pokemon['Speed'])
print("Average Speed:\t\t",avg_speed,end="\n\n")
for type in pd.unique(pokemon['Type 1']):
    print("Type: ",type,end="\t\t")
    temp = np.size(pokemon.loc[((pokemon['Type 1']==type) | (pokemon['Type 2']==type))&(pokemon['Speed']>np.average(pokemon['Speed']))]['Name'])
    print("Count: ",temp)
```

```python
# Lets test our results by checking the flying type
pokemon.loc[((pokemon['Type 1']=="Flying") | (pokemon['Type 2']=="Flying"))&(pokemon['Speed']>avg_speed)]
```

## Sorting

```python
pokemon.sort_values('Name') # Order by name instead of dex number
```

```python
pokemon.sort_values("Speed", ascending=False)[['#','Name','Type 1','Type 2','Speed']] # sort by speed, highest to lowest
```

```python
pokemon.sort_values(["Attack","Sp. Atk"], ascending=False)[['#','Name','Type 1','Type 2','Attack','Sp. Atk']] #Multi attribute sorting
```

```python
pokemon.sort_values(["Attack","Defense"],ascending=[0,1])[["#",'Name','Type 1','Type 2','Attack','Defense']] 
#Attack high to low, defence low to high i.e. glass canons
```
