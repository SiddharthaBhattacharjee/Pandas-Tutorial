## Practice Question:Find number of pokemon with above average speed for each type
(Try it yourself before expanding the solution)

```python
# find the types of pokemon
pd.unique(pokemon['Type 1'])
```

```python
#Solution
import numpy as np
avg_speed = np.average(pokemon['Speed'])
print("Average Speed:\t\t",avg_speed,end="\n\n")
for type in pd.unique(pokemon['Type 1']):
    print("Type: ",type,end="\t\t")
    temp = np.size(pokemon.loc[((pokemon['Type 1']==type) | (pokemon['Type 2']==type))&(pokemon['Speed']>np.average(pokemon['Speed']))]['Name'])
    print("Count: ",temp)
```

```python
# Lets test our results by checking the flying type
pokemon.loc[((pokemon['Type 1']=="Flying") | (pokemon['Type 2']=="Flying"))&(pokemon['Speed']>avg_speed)]
```

## Making Changes to Data

### Creating a new column using existing data

```python
# Find the Base Stat Total (BST) of pokemon
pokemon['BST'] = pokemon['HP'] + pokemon['Attack'] + pokemon['Defense'] + pokemon['Sp. Atk'] + pokemon['Sp. Def'] + pokemon['Speed']
pokemon.head(13)
```

```python
# Now we can sort pokemon from strongest to weekest (The Dataframe retains the new column)
pokemon.sort_values("BST",ascending=False)[0:10]
```

### Dropping a column

```python
pokemon = pokemon.drop(columns=['BST'])
pokemon.head(13)
```

### Another way to create Columns:

```python
pokemon['BST'] = pokemon.iloc[:,4:10].sum(axis=1)# requires less manual work
pokemon.head(13)
```

```python
# Checking if pokemon is a physical or special attacker
pokemon['Atk Type'] = np.where(pokemon['Attack']>pokemon['Sp. Atk'], 'Physical', 'Special')
pokemon[166:175]
```

### Rearranging the data in dataframe:

```python
pokemon = pokemon[['#','Name','Type 1','Type 2','HP','Attack','Defense','Sp. Atk','Sp. Def','Speed','Generation','Legendary','Atk Type','BST']] # swapped attack type and BST
pokemon[166:175]
```

```python
# A smarter way to do it:
cols = list(pokemon.columns.values)
cols
```

```python
pokemon = pokemon[cols[0:10] + cols[-1:-3:-1] + cols[10:12]]
pokemon
```

### Practice Question: Determine weather there are more physical or special attackers

```python
#Get the count of physical attackers
# start by selecting all physical attackers
pokemon[pokemon['Atk Type'] == 'Physical']
```

```python
# Next use count method to get count col wise (vertical)
pokemon[pokemon['Atk Type'] == 'Physical'].count(axis=0)
```

```python
#get a single value of count
pokemon[pokemon['Atk Type'] == 'Physical'].count(axis=0).iloc[0]
```

```python
#assign it to a value and do the same for special attackers
physical_attackers = pokemon[pokemon['Atk Type'] == 'Physical'].count(axis=0).iloc[0]
special_attackers = pokemon[pokemon['Atk Type'] == 'Special'].count(axis=0).iloc[0]
print(f"Physical Attackers: {physical_attackers} \t Special Attackers: {special_attackers}")
print("The dominant attack type is : \t", "Physical" if physical_attackers>special_attackers else "Special")
```
